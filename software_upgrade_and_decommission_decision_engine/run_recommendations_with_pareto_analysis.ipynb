```python
import pandas as pd
from datetime import datetime

# Create a DataFrame from the CSV data
df = pd.read_csv("software_assets_data.csv")

# Function to apply Pareto analysis
def pareto_analysis(item_column, value_column):

    # Group by software version and calculate other software open defects
    aggregated_data = df.groupby(item_column)[value_column].sum().reset_index()

    # Sort data by total defects in descending order
    aggregated_data = aggregated_data.sort_values(by=value_column, ascending=False)

    # Calculate cumulative sum and cumulative percentage
    aggregated_data['Cumulative Sum'] = aggregated_data[value_column].cumsum()
    total_value = aggregated_data[value_column].sum()
    aggregated_data['Cumulative Percentage'] = (aggregated_data['Cumulative Sum'] / total_value) * 100

    # Identify the versions that contribute to 80% of the total defects
    pareto_threshold = 80  # 80% threshold
    aggregated_data['Pareto Contributor'] = aggregated_data['Cumulative Percentage'] <= pareto_threshold

    df['Pareto_Contributor'] = aggregated_data['Pareto Contributor']
    return df

# Function to generate recommendations
def generate_recommendations(df):

    recommendations = []

    current_date = datetime.now()
    item_column="other_software_Version"
    value_column="other_sw_version_defect_count"

    # Group by software version and calculate other software open defects
    aggregated_data = df.groupby(item_column)[value_column].sum().reset_index()

    # Sort data by total defects in descending order
    aggregated_data = aggregated_data.sort_values(by=value_column, ascending=False)

    # Calculate cumulative sum and cumulative percentage
    aggregated_data['Cumulative Sum'] = aggregated_data[value_column].cumsum()
    total_value = aggregated_data[value_column].sum()
    aggregated_data['Cumulative Percentage'] = (aggregated_data['Cumulative Sum'] / total_value) * 100

    # Identify the versions that contribute to 80% of the total defects
    pareto_threshold = 80  # 80% threshold
    aggregated_data['Pareto Contributor'] = aggregated_data['Cumulative Percentage'] <= pareto_threshold

    df['Pareto_Contributor'] = aggregated_data['Pareto Contributor']

    for index, row in df.iterrows():
        software_version = row['software_version']
        software_compatability = row['software_compatbile']
        operate_cost = row['operate_cost']
        upgrade_cost = row['upgrade_cost']
        other_sw_invest_cost = row['other_sw_invest_cost']
        defect_count = row['defect_count']
        other_sw_version_defect_count = row['other_sw_version_defect_count']
        pareto_contributor = row['Pareto_Contributor']

        if (software_compatability == "Yes" or pareto_contributor == "False"):
            action = 'Upgrade'
        else:
            action = 'Decommission'

        recommendations.append({
            'Software Version': software_version,
            'Recommended Action': action
        })

    return recommendations


# Generate recommendations
recommendations = generate_recommendations(df)

# Output the recommendations
for recommendation in recommendations:
    print(f"Software: {recommendation['Software Version']}",
          f"Action: {recommendation['Recommended Action']}")
```
